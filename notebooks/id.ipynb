# Testing ID API SDK

```python
from satvu_api_sdk.sdk import SatVuSDK
import os

sdk = SatVuSDK(
    client_id=os.getenv("SV_CLIENT_ID"),
    client_secret=os.getenv("SV_CLIENT_SECRET"),
    env="qa",
)
```

## User Admin Operations

### Get User Client

```python
sdk.id.get_user_client()
```

### Get User Details

```python
sdk.id.get_user_details()
```

### Edit User Settings

```python
from satvu_api_sdk.services.id.models import NotificationCategory, NotificationConfig, NotificationUpdate, UserSettings

sdk.id.edit_user_settings(UserSettings(notifications=[NotificationUpdate(
    category=NotificationCategory.TASKING,
    settings=[
        NotificationConfig(email=True)
	]
)]))
```

## Webhook Operations

### Create Webhook

```python
from satvu_api_sdk.services.id.models import CoreWebhook

sdk.id.create_webhook(CoreWebhook(
    event_types=[],
    name="Test Webhook",
    url="",
))
```

### List Webhooks

```python
sdk.id.list_webhooks()
```

### Get Webhook Events

```python
sdk.id.get_webhook_events()
```
