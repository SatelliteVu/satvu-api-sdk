# Testing the COS API SDK

```python
from src.satvu_api_sdk.sdk import SatVuSDK
import os

sdk = SatVuSDK(client_id=os.getenv("SV_CLIENT_ID"), client_secret=os.getenv("SV_CLIENT_SECRET"), env="qa")
contract_id = ""
order_id = ""
```

## Query Orders

```python
sdk.cos.query_orders(contract_id)
```

## Get Order Details

```python
sdk.cos.get_order_details(contract_id, order_id)
```

## Edit Order

```python
from cos.models import OrderEditPayload

sdk.cos.edit_order(contract_id, order_id, body=OrderEditPayload(name="My Order"))
```

## Submit Order

```python
from cos.models import OrderSubmissionPayload

sdk.cos.submit_order(contract_id, OrderSubmissionPayload(item_id="20250811T235300000_visual_30_notsat2", name="San Sebastian Bay"))
```

## Download Order

### Redirect

```python
sdk.cos.download_order(contract_id, order_id)
```

### Pre-signed URL

```python
sdk.cos.download_order(contract_id, order_id, redirect=False)
```
